```python
def convert_tuple_string(s):
    name, tuple_part = s.split('=', 1)
    name = name.strip()
    digits = ''.join(filter(str.isdigit, tuple_part))
    return f"{name}_{'_'.join(digits)}"

# Usage
print(convert_tuple_string("mom= [001]"))  # Output: mom_0_0_1
```
