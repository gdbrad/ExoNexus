```python
def output(derivative_coeff_index):
    c, n = derivative_coeff_index
    assert isinstance(n, int) and n >= 0
    d = ["dx", "dy", "dz"]
    ret = []
    num = -1
    while n >= 0:
        num += 1
        n = n - pow(3, num)
    n = n + pow(3, num)
    for _ in range(num):
        ret.append(d[n % 3])
        n = n // 3
    if c == 1:
        pass
    elif c == -1:
        ret.insert(0, "-")
    else:
        ret.insert(0, str(c))
    return "".join(ret)

output([1,4])

from typing import List, Union


class DerivPart:
    """
    Represents a single part of a derivative with a coefficient and a derivative expression.
    """
    def __init__(self, coeff: float, deriv: Union[int, str]) -> None:
        """
        Initializes a DerivPart with a coefficient and a derivative.

        :param coeff: Coefficient of the derivative term (float).
        :param deriv: Derivative identifier, either as an int or a string.
        """
        self.coeff: float = coeff
        self.deriv: Union[int, str] = deriv

    def __repr__(self) -> str:
        """
        Returns a string representation of the DerivPart, formatted as 'coeff * deriv'.
        """
        return f"{self.coeff:.3f} * {self.deriv}"

    def normalize(self, sumsq: float) -> None:
        """
        Normalizes the coefficient based on the sum of squares of the coefficients.

        :param sumsq: Sum of the squares of all coefficients in the derivative.
        """
        self.coeff *= sumsq ** -0.5  # Divide the coefficient by the square root of sumsq


class Deriv:
    """
    Represents a collection of derivative parts, normalizing their coefficients.
    """
    def __init__(self, parts: List[List[Union[float, str]]]) -> None:
        """
        Initializes the Deriv object by normalizing the coefficients of its parts.

        :param parts: A list of lists where each sublist contains a coefficient and a derivative.
                      Example: [[1.0, '1'], [2.0, '2']]
        """
        self.parts: List[DerivPart] = []  # List to hold DerivPart instances
        sumsq = 0.0  # Sum of the squares of the coefficients

        # Loop over the parts to compute the sum of squares and create DerivPart instances
        for part in parts:
            coeff, deriv = part
            sumsq += coeff**2  # Add the square of the coefficient to the sumsq
            self.parts.append(DerivPart(coeff, deriv))  # Create and store DerivPart objects

        # Normalize the coefficients of each DerivPart based on the sum of squares
        for part in self.parts:
            part.normalize(sumsq)

    def __repr__(self) -> str:
        """
        Returns a string representation of the Deriv object by listing all its DerivPart instances.
        """
        return f"{self.parts}"  # List of all parts in the derivative
    
# Deriv([[1, "2 3"], [1, "3 2"]])
Deriv([[1, "1 1"], [-1, "2 2"]])

_naming_scheme = {
    "_": [
        Deriv([[1, ""]]),
    ],
    "d": [
        Deriv([[1, "1"]]),
        Deriv([[1, "2"]]),
        Deriv([[1, "3"]]),
    ],
    "D": [
        Deriv([[1, "2 3"], [1, "3 2"]]),
        Deriv([[1, "3 1"], [1, "1 3"]]),
        Deriv([[1, "1 2"], [1, "2 1"]]),
    ],
    "B": [
        Deriv([[1, "2 3"], [-1, "3 2"]]),
        Deriv([[1, "3 1"], [-1, "1 3"]]),
        Deriv([[1, "1 2"], [-1, "2 1"]]),
    ],
    "E": [
        Deriv([[1, "1 1"], [-1, "2 2"]]),
        Deriv([[-1, "1 1"], [-1, "2 2"], [2, "3 3"]]),
    ],
}

g5 = gamma.gamma[5]
g4 = gamma.gamma[4]

def scheme(name: str):
    assert name in _naming_scheme
    return _naming_scheme[name]

DerivOp = {}


DerivOp['pionxNABLA_T1']
DerivOp['pion_2xNABLA_T1']
DerivOp['pionxD_T2']
DerivOp['pion_2xD_T2']
DerivOp['pionxB_T1']
DerivOp['pion_2xB_T1']

def E_deriv_op(state:str):
    if state.startswith('pi'):
        # _gamma = gamma[0]
        Operator_1 = [1,gamma]
        Operator
```
